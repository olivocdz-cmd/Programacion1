```python
def ingresar_vector(mensaje):
    while True:
        try:
            entrada = input(mensaje).strip()
            vector = [float(x) for x in entrada.split()]
            if not vector:
                print("Error: el vector no puede estar vacío.\n")
                continue
            return vector
        except ValueError:
            print("Error: debe ingresar solo números separados por espacios.\n")

def sumar_vectores(A, B):
    if len(A) != len(B):
        return None, "Error: los vectores deben tener la misma dimensión para sumarse."
    return [a + b for a, b in zip(A, B)], None

def producto_punto(A, B):
    if len(A) != len(B):
        return None, "Error: los vectores deben tener la misma dimensión."
    return sum(a * b for a, b in zip(A, B)), None

def producto_hadamard(A, B):
    if len(A) != len(B):
        return None, "Error: los vectores deben tener la misma dimensión."
    return [a * b for a, b in zip(A, B)], None

def producto_cruz(A, B):
    if len(A) != 3 or len(B) != 3:
        return None, "Error: el producto cruz solo está definido para vectores de dimensión 3."
    x = A[1]*B[2] - A[2]*B[1]
    y = A[2]*B[0] - A[0]*B[2]
    z = A[0]*B[1] - A[1]*B[0]
    return [x, y, z], None

def menu():
    while True:
        print("MENU DE OPCIONES")
        print("1. Sumar vectores")
        print("2. Producto punto")
        print("3. Producto Hadamard")
        print("4. Producto cruz")
        print("5. Salir\n")

        opcion = input("Ingrese una opción (1-5): ").strip()

        if opcion == "5":
            print("Saliendo del programa... ¡Adiós!\n")
            break

        if opcion not in ["1", "2", "3", "4"]:
            print("Opción no válida. Intente de nuevo.\n")
            continue

        v1 = ingresar_vector("Ingrese el vector 1 (números separados por espacios): ")
        v2 = ingresar_vector("Ingrese el vector 2 (números separados por espacios): ")

        resultado, error = None, None

        if opcion == "1":
            resultado, error = sumar_vectores(v1, v2)
            if resultado is not None:
                print(f"Suma: {resultado}\n")
        elif opcion == "2":
            resultado, error = producto_punto(v1, v2)
            if resultado is not None:
                print(f"Producto punto: {resultado}\n")
        elif opcion == "3":
            resultado, error = producto_hadamard(v1, v2)
            if resultado is not None:
                print(f"Producto Hadamard: {resultado}\n")
        elif opcion == "4":
            resultado, error = producto_cruz(v1, v2)
            if resultado is not None:
                print(f"Producto cruz: {resultado}\n")

        if error:
            print(f"{error}\n")

if __name__ == "__main__":
    menu()
```
