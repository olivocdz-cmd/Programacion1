```python
# Ejemplo de conversión de distintos tipos a False

print(0, "->", bool(0))
print(0.0, "->", bool(0.0))
print("", "->", bool(""))
print(list(), "->", bool(list()))
print(tuple(), "->", bool(tuple()))
print(set(), "->", bool(set()))
print(dict(), "->", bool(dict()))
print(None, "->", bool(None))
```

```python
# Ejemplo de conversión de distintos tipos a True

print(9000, "->", bool(9000))
print(3.1415, "->", bool(3.1415))
print("Cadena", "->", bool("Cadena"))
print(list((1, 2, 3, 4)), "->", bool(list((1, 2, 3, 4))))
print(tuple((1, 2, 3, 4)), "->", bool(tuple((1, 2, 3, 4))))
print(set((1, 2, 3, 4)), "->", bool(set((1, 2, 3, 4))))
print(dict([(1,2), (3,4), (5,6)]), "->", bool(dict([(1,1), (2,2), (3,3)])))
```

```python
# Ejemplo if-else

nota = float(input("Introduzca su nota: "))
if nota < 6:
  print("No aprobaste, sigue intentando")
else:
  print("Aprobaste, ¡felicitaciones!")
```

```python
#Ejemplo if-elif-else

nota = float(input("Introduzca su nota: "))
if 0 <= nota and nota < 4:
  print("Muy bajo, reprobado")
elif 4 <= nota and nota < 6:
  print("Regular")
elif 6 <= nota and nota < 8:
  print("Bueno")
elif 8 <= nota and nota <= 10:
  print("Excelente")
else:
  print("Valor inválido")
```

```python
# Ejemplo while

contador = 5
while contador != 0:
  contador = contador - 1
  print("El valor actual del contador es", contador)
```

```python
# Ejemplo básico del for

for iterador in [1, 2, 3, 4, 5, 6]:
  print(iterador)
```

```python
# for sobre estructuras de datos

lista = list((2, 1, 0, 3, 4))
tupla = tuple((2, 1, 0, 3, 4))
conjunto = set((2, 1, 0, 3, 4))
diccionario = dict(((2,"Dos"), (1, "Uno"), (0, "Cero"), (3, "Tres"), (4, "Cuatro")))

for iterador in lista:
  print(iterador)
```

```python
# Acceso por índices a una lista con un for

lista = ["Vamos", "a", "acceder", "a", "esta", "lista", "por", "índices", "de", "forma", "ordenada"]

for indice in range(len(lista)):
  print("Indice:", indice,"Elemento:", lista[indice])
```

```python
# for para visualizar el efecto de las sentencias break y continue

for numero in range(8):
  print(numero)
```

```python
# Visualización del efecto de la sentencia break

for numero in range(7):
  if numero == 4:
    break
  print(numero)
```

```python
# Visualización del efecto de la sentencia continue

for numero in range(7):
  if numero == 4:
    continue
  print(numero)
```

```python
# Uso del else en un while

intentos = 4
while intentos > 0:
  if input(">>> Ingrese la clave:") == "Clave123":
    print("¡Acceso concedido!")
    break
  intentos = intentos - 1
  print("Clave incorrecta")
else:
  print("Se agotaron los intentos :(")
```
