```python
def idnumero(m): # Nombre de la función
  try: # Le indica a la función que intente ejecutar el programa aunque no se cumpla una condición.
    m = int(m)
    if isinstance(m, int) and m >= 0:
      for n in range(0, (m+1)):
        if n == 1 or n == 0:
          print("El número es ", n)
        else:
          for x in range(2, n):
            if n % x == 0:
              print(n, 'es igual a', x, '*', n // x)
              break
            else: # Este else corresponde al for, no al if
              print(n, 'es un número primo')
    else:
      print("El número no es natural")
  except ValueError: # Le indica al código que iverte un error de valor, es decir, que si no se cumplen las copndiciones
                    # ofrezca un error e indique que debe intrudicir un parámetro permitido.
    print("El valor ingresado no es un número")
```

```python
def contador_vocales1(x): # Definición de la función y su argumento
  vminus = "aeiou" # Se asignan las variables
  vmayus = "AEIOU"
  cminus = "bcdfghjklmnñpqrstvwxyz"
  cmayus = "BCDFGHJKLMNÑPQRSTVWXYZ"
  numeros = "0123456789"
  espacio = " "

  contador0 = 0 # Variables vacías para aignar un valor posteriormente
  contador1 = 0
  contador2 = 0
  contador3 = 0
  contador4 = 0
  contador5 = 0
  contador6 = 0

  for letra in x: # Avalua a cada elemento dentro de la cadena o argumento de la función
    if letra in vmayus:  # Cuenta las vovales mayúsculas
      contador0 += 1
    elif letra in vminus:  # Cuenta las vovales minúsculas
      contador1 += 1
    elif letra in cmayus:
      contador2 += 1
    elif letra in cminus:
      contador3 += 1
    elif letra in numeros:
      contador4 += 1
    elif letra in espacio:
      contador5 += 1
    else:
      contador6 += 1
  return({'Cantidad de vocales mayúsculas' : contador0, # Establece un diccionario como salida
        "Cantidad de vocales minúsculas:": contador1,
        "Cantidad de consonantes mayúsculas:": contador2,
        "Cantidad de consonantes minúsculas:": contador3,
        "Cantidad de números:": contador4,
        "Cantidad de espacios:": contador5,
        "Cantidad de otros caracteres:": contador6})
```

```python
palabrita = input("Ingrese una palabra: ")
contador_vocales1(palabrita)
```
