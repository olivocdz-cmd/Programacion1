```python
# Listas

lista = [29, True, 3.1415, "El número de Euler sí que mola", "La M.O.D.A"]

print(lista)
print(lista[-1])
print(lista[1:4])

lista[2] = "Yo soy tu padre"

lista[2] = [3, 2, 1]
print(lista)

print(len(lista))
```

```python
lista_nueva = [1, 2, 3, 4, 5, 6, 10 ,9]

lista_nueva.append(3)

print(lista_nueva)

print(lista_nueva.count(3))

print(lista_nueva.index(4))

lista_nueva.remove(3)

print(lista_nueva)
```

```python
# Tuplas

tupla = ("¿La tierra es plana?", True, False, "TOP")
print(tupla)

print(tupla[0])
print(tupla[1])
print(tupla[2])
print(tupla[3])

print(tupla.count(True))
print(tupla.index(False))

print((1))
print((1, 3))
```

```python
# Conjunto

print(set())
print(set([5, 2, 5, 1, 1.5, 3, 2.5]))
print(set((5, 2, 5, 1, 1.5, 3, 2.5)))
print(set("52511.538"))

conjunto = set([2, 3, 3, 4, 7, 8])
conjunto_2 = set([5, 3, 5, 6, 9, 10])
conjunto_3 = set([4, 2, 11])

print(conjunto)
print(conjunto_2)
print(conjunto_3)

conjunto.add(1)
print(conjunto)

conjunto.remove(1)
print(conjunto)

print(conjunto.intersection(conjunto_2))
print(conjunto_2.issubset(conjunto))
print(conjunto_3.issubset(conjunto))
```

```python
# Diccionario

diccionario = {1: "Uno", 2: "Dos", 4: "Cuatro", 5: "Cinco"}
diccionario[3] = "Tres"
print(diccionario)

dict_lista_tuplas = dict([(1, "Uno"), (2, "Dos"), (3, "Tres"), (4, "Cuatro"), (5, "Cinco")])
print(dict_lista_tuplas)

dict_lista_string = dict(Uno = 1, Dos = 2, Tres = 3, Cuatro = 4, Cinco = 5)
print(dict_lista_string)

dict_tipos = {1: "integer",
              2.2: "float",
              "texto": "string",
              (1, 2): "tupla",
              True: "boolean",
              3+2j: "complejo"}

print(dict_tipos)

dict_repeticion = {1: "Primero", 1: "Último", 2: "Segundo", 2: "Final"}
print(dict_repeticion)

print(diccionario,
      diccionario.keys(),
      diccionario.values(),
      diccionario.items())

claves = diccionario.values()
print(claves)
diccionario[1] = "One"
print(claves)
diccionario.pop(2)
print(diccionario)
```
