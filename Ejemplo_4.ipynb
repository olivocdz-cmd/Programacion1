```python
def ingresar_matriz():
    while True:
        try:
            filas = int(input("Ingrese el número de filas: ").strip())
            columnas = int(input("Ingrese el número de columnas: ").strip())
            if filas <= 0 or columnas <= 0:
                print("Error: Las dimensiones deben ser mayores a 0.\n")
                continue
            break
        except ValueError:
            print("Error: Debe ingresar solo números enteros.\n")

    matriz = []
    for i in range(filas):
        while True:
            try:
                entrada = input(f"Ingrese la fila {i+1} separada por espacios: ").strip()
                fila = list(map(float, entrada.split()))
                if len(fila) != columnas:
                    print(f"Error: Debe ingresar exactamente {columnas} valores.\n")
                    continue
                matriz.append(fila)
                break
            except ValueError:
                print("Error: Debe ingresar solo números.\n")
    return matriz

def transponer_matriz(A):
    try:
        if not A or not A[0]:
            print("Error: La matriz no puede estar vacía.\n")
            return None

        filas = len(A)
        columnas = len(A[0])

        resultado = []

        for j in range(columnas):
            nueva_fila = []
            for i in range(filas):
                nueva_fila.append(A[i][j])
            resultado.append(nueva_fila)

        return resultado

    except (TypeError, ValueError):
        print("Error: Datos inválidos.\n")
        return None

def determinante(A):
    n = len(A)

    if n == 1:
        return A[0][0]

    if n == 2:
        return A[0][0]*A[1][1] - A[0][1]*A[1][0]

    det = 0
    for j in range(n):
        det += ((-1) ** j) * A[0][j] * determinante(
            [fila[:j] + fila[j+1:] for fila in A[1:]]
        )
    return det

def cofactor(A, i, j):
    signo = (-1) ** (i + j)

    submatriz = [
        fila[:j] + fila[j+1:]
        for k, fila in enumerate(A) if k != i
    ]

    return signo * determinante(submatriz)

def cofactor_matriz(A):
    if not A or not A[0]:
        print("Error: La matriz no puede estar vacía.\n")
        return None

    filas = len(A)
    columnas = len(A[0])

    if filas != columnas:
        print("Error: La matriz debe ser cuadrada.\n")
        return None

    for fila in A:
        if len(fila) != columnas:
            print("Error: La matriz debe ser cuadrada.\n")
            return None

    resultado = []
    for i in range(filas):
        fila_cofactor = []
        for j in range(columnas):
            fila_cofactor.append(cofactor(A, i, j))
        resultado.append(fila_cofactor)
    return resultado

def vectorizar(A):
    if not A or not A[0]:
        print("Error: La matriz no puede estar vacía.\n")
        return None

    resultado = []
    for i in range(len(A)):
        for j in range(len(A[0])):
            resultado.append([A[i][j]])
    return resultado

```

```python
def menu_transformaciones():
    while True:
        print("\nMenú de opciones:")
        print("Ingrese la operación que desea realizar:")
        print("1. Transponer una matriz")
        print("2. Determinar la matriz de cofactores")
        print("3. Vectorizar por filas una matriz")
        print("4. Salir\n")
        opcion = input("Seleccione una opción: ").strip()

        if opcion == "4":
            print("Saliendo del programa.\n")
            break

        if opcion not in ["1", "2", "3"]:
            print("Opción no válida. Intente de nuevo.\n")
            continue

        print("\nIngrese la matriz:")
        A = ingresar_matriz()

        if A is None:
            continue

        if opcion == "1":
            resultado = transponer_matriz(A)
            if resultado is not None:
                print("\nLa matriz transpuesta es:")
                for fila in resultado:
                    print(fila)
                print()
        elif opcion == "2":
            resultado = cofactor_matriz(A)
            if resultado is not None:
                print("\nLa matriz de cofactores es:")
                for fila in resultado:
                    print(fila)
                print()
        elif opcion == "3":
            resultado = vectorizar(A)
            if resultado is not None:
                print("\nLa matriz vectorizada por filas es:")
                for fila in resultado:
                    print(fila)
                print()
```

```python
menu_transformaciones()
```
