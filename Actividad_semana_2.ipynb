```python
animales = ['Tigre', 'Jaguar', 'Ballena', 'Oso', 'Mosquito', 'Cuervo', 'Tejón']
print('De lista a tupla: ', tuple(animales))
print('De lista a a conjunto: ', set(animales))
print('De lista a diccionario: ', dict(enumerate(animales)))
```

```python
ciudades = ('Valencia', 'Madrid', 'Alicante', 'Sevilla', 'Pamplona', 'Mallorca')
print('De tupla a lista: ', list(ciudades))
print('De tupla a conjunto: ', set(ciudades))
print('De tupla a diccionario: ', dict(enumerate(ciudades)))
```

```python
idiomas = {'Español', 'Inglés', 'Francés', 'Latín', 'Griego', 'Mandarín'}
print('De conjunto a lista: ', list(idiomas))
print('De conjunto a tupla: ', tuple(idiomas))
print('De conjunto a diccionario: ', dict(enumerate(idiomas)))
```
